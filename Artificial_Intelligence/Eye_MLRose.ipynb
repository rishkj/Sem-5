```python
from datetime import datetime
startTime = datetime.now()

l = []
def generateColumns(start, end):
    for i in range(start, end+1):
        l.extend([str(i)+'X', str(i)+'Y'])
    return l

eyes = generateColumns(1, 12)

# reading in the csv as a dataframe
# !pip3 install pandas
import pandas as pd
df = pd.read_csv('EYES.csv')

# selecting the features and target
X = df[eyes]
y = df['truth_value']

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.10, random_state = 42)

# Data Normalization
from sklearn.preprocessing import StandardScaler as SC
sc = SC()
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)

import numpy as np
X_train, y_train, X_test, y_test = np.array(X_train), np.array(y_train), np.array(X_test), np.array(y_test)

import mlrose

nn_model1 = mlrose.NeuralNetwork(hidden_nodes = [8,8], activation = 'relu',
                                 algorithm = 'simulated_annealing', max_iters = 1000,
                                 bias = True, is_classifier = True, learning_rate = 0.0001,
                                 early_stopping = True, clip_max = 5, max_attempts = 100,
				 random_state = 3)

nn_model1.fit(X_train, y_train)

from sklearn.metrics import accuracy_score
	
# Predict labels for train set and assess accuracy
y_train_pred = nn_model1.predict(X_train)

y_train_accuracy = accuracy_score(y_train, y_train_pred)

print('Training accuracy: ', y_train_accuracy)

# Predict labels for test set and assess accuracy
y_test_pred = nn_model1.predict(X_test)

y_test_accuracy = accuracy_score(y_test, y_test_pred)

print('Test accuracy: ', y_test_accuracy)

df_results = pd.DataFrame()
df_results['Actual label'] = y_test
df_results['Predicted value'] = y_test_pred
#df_results['Predicted label'] = y_pred_labels

df_results1 = pd.DataFrame()
df_results1['Actual label'] = y_train
df_results1['Predicted value'] = y_train_pred
df_results1.to_csv('Results_MLRose.csv')

print(df_results)

print("\n")
print(df_results1)


print("\n")
print("Execution time in seconds = ", datetime.now() - startTime)
```
